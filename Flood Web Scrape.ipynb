# Import Packages

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
import os
import time
import random

stations = pd.read_csv('Hydro_to_Climate_Mapping.csv')
```

# Web Scrape

### Get Climate Data Main Page

```python
page_url = 'https://dd.weather.gc.ca/today/climate/observations/daily/csv/BC/'

# Get and read page data
page = requests.get(page_url)
soup = BeautifulSoup(page.content, 'html.parser')

# Store main page data
with open("bc_climate.html", 'w', encoding='utf-8') as f:
    f.write(str(soup))
```

### Get Individual Climate Data Files

```python
# Read Province Climate File
bc_climate = open("bc_climate.html", 'r', encoding='utf-8').read()
soup = BeautifulSoup(bc_climate, 'html.parser')

# Get all files on the page
all_files = soup.find_all('a', href=True)

# Set year range
last_year = 2026
first_year = 2018

# Set List of Climate Stations to scrape
climate_ids = stations['Climate ID'].unique()

# Set regex to match years and ids
year_regex = re.compile(r'_(\d{4})_')
id_regex = re.compile(r'BC_([a-zA-Z0-9]+)_\d{4}')

# Filter all files to get the subset of files to scrape
for file in all_files:
    year = year_regex.search(file['href'])
    climate_id = id_regex.search(file['href'])
    if (year) and (first_year <= int(year.group(1)) <= last_year) and (climate_id.group(1) in climate_ids):
        data_file = 'https://dd.weather.gc.ca/today/climate/observations/daily/csv/BC/' + file['href']
        df = pd.read_csv(data_file, encoding='utf-8', encoding_errors='replace')
        df.to_csv('climate_files/' + file['href'], index=False)
        print(file['href'])
        sec = random.randint(5, 20)
        time.sleep(sec)
```

### Get Hydro Data Main Page

```python
page_url = 'https://collaboration.cmc.ec.gc.ca/cmc/hydrometrics/www/UnitValueData/Stage/corrected/08/'

# Get and read page data
page = requests.get(page_url)
soup = BeautifulSoup(page.content, 'html.parser')

# Store main page data
with open("bc_hydro.html", 'w', encoding='utf-8') as f:
    f.write(str(soup))

# Manually downloaded 7 files from different pages
# '07FA004', '07FB003', '07FB004', '07FC001', '07FD002', '07FD007', '10CD001'
```

### Get Individual Hydro Data Files

```python
# Read Province Hydro File
bc_hydro = open("bc_hydro.html", 'r', encoding='utf-8').read()
soup = BeautifulSoup(bc_hydro, 'html.parser')

all_files = soup.find_all('a', href=True)

id_regex = re.compile(r'@([A-Z0-9]+)\.')

hydro_ids = stations['Hydro ID'].unique()

# Filter all files to get the subset of files to scrape
for file in all_files:
    hydro_id = id_regex.search(file['href'])
    if (hydro_id) and (hydro_id.group(1) in hydro_ids):
        data_file = 'https://collaboration.cmc.ec.gc.ca/cmc/hydrometrics/www/UnitValueData/Stage/corrected/08/' + file['href']
        df = pd.read_csv(data_file, encoding='utf-8', encoding_errors='replace', skiprows=14, on_bad_lines='warn', engine='python')
        df.to_csv('hydro_files/' + file['href'], index=False)
        print(file['href'])
        sec = random.randint(5, 30)
        time.sleep(sec)
```
