# Reading and filtering climate station file

```python
import pandas as pd

climate_stations = pd.read_csv('data/climate_station_list.csv')
climate_stations = climate_stations[climate_stations['DLY First Year'] <= 2015]
climate_stations = climate_stations[climate_stations['DLY Last Year'] >= 2025]
climate_stations = climate_stations[climate_stations['Province'] == 'BRITISH COLUMBIA']
```

# EDA on filtered climate station dataframe

```python
print(climate_stations.head())

print(climate_stations.info())

print(climate_stations.describe())

print(climate_stations.isna().sum())
```

```python
# Convert all column names to lowercase
climate_stations.columns = climate_stations.columns.str.lower()

# Get the unique combinations
unique_combos = climate_stations[['latitude', 'longitude']].value_counts()

# Total number of unique combinations
print(f"Unique combinations: {len(unique_combos)}")

# Total number of rows in the dataframe
print(f"Total rows: {len(climate_stations)}")
```

# Dropping duplicate coordinate pairs and saving unique coordinates to json file

```python
import json

unique_coords = climate_stations[['latitude', 'longitude']].drop_duplicates()
coords_dict = unique_coords.to_dict(orient='records')
with open('unique_coordinates.json', 'w') as f:
    json.dump(coords_dict, f)
```
